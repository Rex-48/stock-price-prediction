## @Downloading dataset

```python
## downloading dataset in the Like a chad 

import kagglehub
import shutil
import os

dataset_handle = "dgawlik/nyse"

# Step 1: Summon the mighty Kaggle servers to drop the dataset into cache
download_path = kagglehub.dataset_download(dataset_handle)

print("Dataset downloaded to cache:", download_path)

destination = os.path.abspath("../data/nyse")
os.makedirs(destination, exist_ok=True)

# Copying files manually like it’s 2005 because kagglehub refuses to just save them there
for file in os.listdir(download_path):
    src = os.path.join(download_path, file)
    dst = os.path.join(destination, file)
    if os.path.isfile(src):
        shutil.copy2(src, dst)

print("Dataset copied to:", destination)

import pandas as pd

# Finally, the good stuff — data!
prices = pd.read_csv(os.path.join(destination, "prices.csv"))
print(prices.head())
```

```python
df = df[df['symbol'] == 'AAP']  # Laser-focusing on one poor company's data

# Kick out the columns that can't do math before asking for correlations
cormap = df.select_dtypes(include='number').corr()

fig, ax = plt.subplots(figsize=(5, 5))
sns.heatmap(cormap, annot=True)
plt.show()
```

```python
def get_corelated_col(cor_dat, threshold):
    # Only the drama queens above this threshold get invited
    feature = []
    value = []

    for i, index in enumerate(cor_dat.index):
        if abs(cor_dat[index]) > threshold:
            feature.append(index)
            value.append(cor_dat[index])

    # Making a nice little table for the popular columns
    df = pd.DataFrame(data=value, index=feature, columns=['corr value'])
    return df
```

```python
top_corelated_values = get_corelated_col(cormap['close'], 0.60)
top_corelated_values
```

> Looks like all colums except _volume_ are ***highly co-related***. Using them for predictions.

```python
df = df[top_corelated_values.index]
df.head()
```

```python
df.shape
```

```python
sns.pairplot(df)
plt.tight_layout()
```

> Since other parameters have linear relationship with __close__, we are using some linear models fore prediction

```python
X = df.drop(['close'], axis=1)
y = df['close']
```

> Since range of data in different columns veries significantly we need to scale the independent variable i.e. X. For this we will use Min-Max Scaling.

```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

scaler = MinMaxScaler()
# Forcing all columns to obey the 0-1 dictatorship
X = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)
X.head()
```

```python
from sklearn.model_selection import train_test_split

# Let's split the chaos into a training playground and a testing lab
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# To keep track of how well we pretend to predict
Acc = []
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression

# Summoning the almighty Linear Regression to try and predict the future
model_1 = LinearRegression()
model_1.fit(X_train, y_train)
```

```python
# Asking Linear Regression to spill the future
y_pred_1 = model_1.predict(X_test)

# Comparing reality vs. the model's crystal ball
pred_df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred_1})
pred_df.head()
```

```python
from sklearn.metrics import r2_score

# How much does our crystal ball suck… I mean, predict?
score = r2_score(y_test, y_pred_1)
print(f"Accuracy score of the predictions: {score}")

# Stashing the glory (or shame) for later
Acc.append(score)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 8))
plt.ylabel('Close Price', fontsize=16)

# Actual vs model's best guess
plt.plot(pred_df['Actual'], color='blue')
plt.plot(pred_df['Predicted'], color='orange')

plt.legend(['Actual Value', 'Predictions'])
plt.title("Actual vs Predicted Close Prices (Linear Regression trying its best)")
plt.show()
```

## ANN

```python
from keras.models import Sequential
from keras.layers import Dense

def regressor(inp_dim):
    # Summoning the mighty feedforward neural network to try predicting the unpredictable
    model = Sequential()
    
    model.add(Dense(20, input_dim=inp_dim, kernel_initializer='normal', activation='relu'))
    model.add(Dense(25, kernel_initializer='normal', activation='relu'))
    model.add(Dense(10, kernel_initializer='normal', activation='relu'))
    model.add(Dense(1, kernel_initializer='normal'))  # One output to rule them all
    
    model.compile(loss='mean_squared_error', optimizer='adam')
    
    return model

```

```python
# the polishing the crystal ball 

model_2 = regressor(inp_dim=3)
model_2.fit(X_train, y_train, epochs=70, validation_split=0.2)
```

```python
# Prediction

y_pred_2 = model_2.predict(X_test)
```

```python
pred_df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred_2.flatten()})
pred_df.head()
```

```python
# Measure the Accuracy Score

from sklearn.metrics import r2_score
 
print("Accuracy score of the predictions: {0}".format(r2_score(y_test, y_pred_2)))
Acc.append(r2_score(y_test, y_pred_2))
```

```python
plt.figure(figsize=(8,8))
plt.ylabel('Close Price', fontsize=16)
plt.plot(pred_df)
plt.legend(['Actual Value', 'Predictions'])
plt.show()
```

## CNN

```python
X_train = np.array(X_train).reshape(X_train.shape[0], X_train.shape[1], 1)
X_test = np.array(X_test).reshape(X_test.shape[0], X_test.shape[1], 1)
```

```python
from tensorflow.keras import Sequential,utils
from tensorflow.keras.layers import Flatten, Dense, Conv1D, MaxPool1D, Dropout

def reg():
    
    model = Sequential()
    
    model.add(Conv1D(32, kernel_size=(3,), padding='same', activation='relu', input_shape = (X_train.shape[1],1)))
    model.add(Conv1D(64, kernel_size=(3,), padding='same', activation='relu'))
    model.add(Conv1D(128, kernel_size=(5,), padding='same', activation='relu'))
    
    model.add(Flatten())
    
    model.add(Dense(50, activation='relu'))
    model.add(Dense(20, activation='relu'))
    model.add(Dense(units = 1))
    
    model.compile(loss='mean_squared_error', optimizer='adam')
    
    return model
```

```python
# Model Training

model_3 = reg()
model_3.fit(X_train, y_train, epochs=100, validation_split=0.2)
```

```python
# Prediction

y_pred_3 = model_3.predict(X_test)
```

```python
# Measure the Accuracy Score

from sklearn.metrics import r2_score
 
print("Accuracy score of the predictions: {0}".format(r2_score(y_test, y_pred_3)))
Acc.append(r2_score(y_test, y_pred_3))
```

```python
plt.figure(figsize=(8,8))
plt.ylabel('Close Price', fontsize=16)
plt.plot(pred_df)
plt.legend(['Actual Value', 'Predictions'])
plt.show()
```

```python
plt.plot(range(3), Acc, color='green', linestyle='dashed', linewidth = 3, 
         marker='o', markerfacecolor='blue', markersize=12) 
plt.ylabel('Acc')
plt.xlabel('Models')
plt.title("Accuracies")
plt.xticks(range(3), ['Linear Regression', 'ANN', 'CNN'])
plt.show()
```
